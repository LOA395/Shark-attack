# Proyecto de Limpieza de Datos de Ataques de Tiburones

Este proyecto tiene como objetivo limpiar y estandarizar un conjunto de datos relacionado con ataques de tiburones a nivel mundial. El conjunto de datos original contiene información sobre los incidentes, incluyendo la fecha, la ubicación, la actividad realizada durante el ataque, y otros detalles. A través de este proceso de limpieza, buscamos mejorar la calidad de los datos para que puedan ser analizados de manera más efectiva.

### Pasos del Proyecto
El proceso de limpieza y estandarización de datos se llevó a cabo mediante los siguientes pasos:

1. **Carga y verificación de los datos**
Primero, cargamos el archivo de Excel original y verificamos su contenido para asegurarnos de que estaba correctamente formateado y accesible. Esto incluyó la configuración de pandas para mostrar todas las columnas y la inspección inicial del DataFrame, para identificar rápidamente posibles problemas, como valores faltantes o columnas irrelevantes.
2. **Filtrado por año**
Eliminamos registros anteriores al año 1700, ya que los datos de esas fechas eran menos fiables, con muchos valores faltantes y errores de formato. Al centrarnos en datos desde 1700 en adelante, garantizamos una mayor calidad y relevancia en el conjunto de datos.
3. **Tratamiento de valores nulos**
Para mejorar la calidad del conjunto de datos, eliminamos las columnas que tenían más del 50% de valores faltantes y las filas con más de cinco valores faltantes. Este paso ayuda a reducir el ruido y a concentrarse en datos más completos.
4. **Eliminación de columnas irrelevantes**
Eliminamos columnas que no aportaban información útil para el estudio, como 'pdf', 'href', y 'Case Number', entre otras. También eliminamos otras columnas como 'Injury' y 'Species', que aunque podrían ser interesantes, tenían demasiados valores nulos o inconsistencias.
5. **Renombrar columnas**
Renombramos la columna 'Unnamed: 11' a 'Deceased', para reflejar mejor el contenido de la columna, que indica si la persona falleció en el ataque.
6. **Estandarización de columnas clave**
Realizamos la estandarización de los valores en varias columnas clave:

    - País: Estandarizamos los nombres de países utilizando la librería pycountry, y aplicamos un mapeo manual para corregir nombres inconsistentes o mal escritos.
    - Estado: Estandarizamos los nombres de los estados, convirtiéndolos a mayúsculas y eliminando espacios en blanco innecesarios.
    - Actividad: Simplificamos y agrupamos las actividades relacionadas con los ataques mediante expresiones regulares, reduciendo la variedad a categorías clave como 'swimming', 'surfing', 'fishing', etc.
    - Sexo: Estandarizamos los valores a 'M', 'F' y 'Unknown', corrigiendo errores y manejando los valores faltantes.
    - Edad: Convertimos la columna de edad a formato numérico, eliminando entradas no numéricas para asegurar la consistencia.
    - Mortalidad (Deceased): Estandarizamos los valores a 'Y', 'N' y 'Unknown', corrigiendo errores tipográficos y valores inconsistentes.

7. **Estandarización de fechas**
Estandarizamos las fechas para que siguieran un formato consistente, lo que incluyó eliminar prefijos como 'Reported-', 'Late-', y 'Early-', y extraer el mes y el año en columnas separadas.
8. **Creación de columnas adicionales**
Para facilitar el análisis, añadimos columnas adicionales como:

    - Década: Para agrupar los datos por décadas.
    - Estación: Basada en el mes y el país, determinamos la estación (verano, invierno, etc.) en la que ocurrió el ataque.


9. **Eliminación de duplicados**
Eliminamos registros duplicados para asegurar que cada incidente esté representado solo una vez, y restablecimos el índice del DataFrame.
10. **Guardado del archivo limpio**
Finalmente, guardamos los datos limpios en un archivo CSV para su posterior análisis.
